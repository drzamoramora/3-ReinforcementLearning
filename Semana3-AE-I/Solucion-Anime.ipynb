## Solucion Generados de Anime

```python
# Cargar datos

import tensorflow as tf
from tensorflow import keras
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
from keras_preprocessing.image import ImageDataGenerator

ENTRENAMIENTO_DIR = "C:/Users/XPC/Desktop/Datasets/anime-faces/"
SIZE = 64

# configuracion de entrenamiento
datagen = ImageDataGenerator(rescale = 1./255)

# generador
gen = datagen.flow_from_directory(
    ENTRENAMIENTO_DIR,
    target_size=(SIZE,SIZE),
    batch_size=16,
    class_mode='input'
)
```

### Arquitectura AE-Convolucional

```python
tf.random.set_seed(42)
np.random.seed(42)

# funcion de exactitud binaria
def rounded_accuracy(y_true, y_pred):
    return keras.metrics.binary_accuracy(tf.round(y_true), tf.round(y_pred))

# encoder
conv_encoder = keras.models.Sequential([
    keras.layers.Input((SIZE,SIZE,3), name="enc_input"),
    keras.layers.Conv2D(16, kernel_size=3, padding="SAME", activation="selu"),
    keras.layers.MaxPool2D(pool_size=2),
    keras.layers.Conv2D(32, kernel_size=3, padding="SAME", activation="selu"),
    keras.layers.MaxPool2D(pool_size=2),
    keras.layers.Conv2D(64, kernel_size=3, padding="SAME", activation="selu"),
    keras.layers.MaxPool2D(pool_size=2)
])

conv_encoder.summary()

conv_encoder.layers
```

```python
# decoder
conv_decoder = keras.models.Sequential([
    keras.layers.Conv2DTranspose(32, kernel_size=3, strides=2, padding="SAME", activation="selu",
                                 input_shape=[8, 8, 64]),
    keras.layers.Conv2DTranspose(16, kernel_size=3, strides=2, padding="SAME", activation="selu"),
    keras.layers.Conv2DTranspose(3, kernel_size=3, strides=2, padding="SAME", activation="sigmoid"),
    keras.layers.Reshape([64, 64,3])
])

conv_decoder.summary()
```

```python
# autoencoder
conv_ae = keras.models.Sequential([conv_encoder, conv_decoder])


# compilar y entrenar
conv_ae.compile(loss="binary_crossentropy", optimizer=keras.optimizers.SGD(lr=1.0),
                metrics=[rounded_accuracy])

history = conv_ae.fit_generator(gen,epochs=20,shuffle=True)
```

### Predecir una imagen del Dataset (datagen)

```python
from skimage import io

def imshow(image_RGB):
    io.imshow(image_RGB)
    io.show()
    
x,y = gen.next()
c = conv_ae.predict(x)[0]
imshow(x[0])
imshow(c)
```
