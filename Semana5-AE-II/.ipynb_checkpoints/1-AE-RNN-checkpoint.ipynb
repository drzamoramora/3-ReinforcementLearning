## RNN AE

En esta semana continuaremos con los auto-encoders (AE). Vamos a revisar una modificación bastante particular del stacked AE que vimos la semana anterior. Esta vez aprenderemos que también podemos generar AE con series temporales. Recordemos que para procesar series temporales podemos utilizar RNN. 

Normalmente un RNN AE se compone de del 1) Encoder: una RNN de secuencia-a-vector y el 2) Decoder: realiza la operación inversa convirtiendo el vector nuevamente a una secuencia.

```python
import tensorflow as tf
from tensorflow import keras
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
```

### Cargar las imagenes del Fashion MNIST

```python
# CARGAR FASHION MNIST
(X_train_full, y_train_full), (X_test, y_test) = keras.datasets.fashion_mnist.load_data()
X_train_full = X_train_full.astype(np.float32) / 255
X_test = X_test.astype(np.float32) / 255
X_train, X_valid = X_train_full[:-5000], X_train_full[-5000:]
y_train, y_valid = y_train_full[:-5000], y_train_full[-5000:]

X_train.shape
```

### Funciones de Utileria - Accuracy

```python
def rounded_accuracy(y_true, y_pred):
    return keras.metrics.binary_accuracy(tf.round(y_true), tf.round(y_pred))
```

### AE RNN

Nuestro Encoder recibe secuencias de cualquier longitud con 28 dimensiones por step/time. Lo que pasa con el Fashion MNIST, es que va a procesar cada imagen como una serie temporal, procesando arreglos de 28 pixeles cada step.  

El Decoder utiliza una primera capa llamada RepeatVector, que permite recibir un vector de 28 procesarlo como una secuencia. El resto, es simplemente el inversión del encoder.

```python
# ENCODER
recurrent_encoder = keras.models.Sequential([
    keras.layers.LSTM(100, return_sequences=True, input_shape=[28, 28]),
    keras.layers.LSTM(30)
])

# DECODER
recurrent_decoder = keras.models.Sequential([
    keras.layers.RepeatVector(28, input_shape=[30]),
    keras.layers.LSTM(100, return_sequences=True),
    keras.layers.TimeDistributed(keras.layers.Dense(28, activation="sigmoid"))
])

# RNN AE
recurrent_ae = keras.models.Sequential([recurrent_encoder, recurrent_decoder])

# COMPILE & TRAIN
recurrent_ae.compile(loss="binary_crossentropy", optimizer=keras.optimizers.SGD(0.1),
                     metrics=[rounded_accuracy])
history = recurrent_ae.fit(X_train, X_train, epochs=10, validation_data=(X_valid, X_valid))
```

### Funciones de Utileria (plot & display)

```python
def plot_image(image):
    plt.imshow(image, cmap="binary")
    plt.axis("off")
    
def show_reconstructions(model, images=X_valid, n_images=5):
    reconstructions = model.predict(images[:n_images])
    fig = plt.figure(figsize=(n_images * 1.5, 3))
    for image_index in range(n_images):
        plt.subplot(2, n_images, 1 + image_index)
        plot_image(images[image_index])
        plt.subplot(2, n_images, 1 + n_images + image_index)
        plot_image(reconstructions[image_index])
```

### Utilizar el RNN AE

Las imágenes se ven como series temporales…

```python
show_reconstructions(recurrent_ae)
plt.show()
```
