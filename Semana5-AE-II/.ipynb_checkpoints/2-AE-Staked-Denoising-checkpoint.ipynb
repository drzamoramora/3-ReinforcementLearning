## Stacked Denoising AE

Otra forma de obligar al codificador automático a aprender funciones útiles es agregar ruido a sus entradas, entrenándolo para recuperar las entradas originales, libres de ruido. Esto evita autoencoder de copiar trivialmente sus entradas a sus salidas, por lo que termina encontrando patrones en los datos.

Vamos a generar ruido mediante 2 formas: con GaussianNoise y con Dropout.

```python
import tensorflow as tf
from tensorflow import keras
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
```

### Cargar las imagenes del Fashion MNIST

```python
# CARGAR FASHION MNIST
(X_train_full, y_train_full), (X_test, y_test) = keras.datasets.fashion_mnist.load_data()
X_train_full = X_train_full.astype(np.float32) / 255
X_test = X_test.astype(np.float32) / 255
X_train, X_valid = X_train_full[:-5000], X_train_full[-5000:]
y_train, y_valid = y_train_full[:-5000], y_train_full[-5000:]

X_train.shape
```

### Funciones de Utileria - Accuracy

```python
def rounded_accuracy(y_true, y_pred):
    return keras.metrics.binary_accuracy(tf.round(y_true), tf.round(y_pred))
```

**keras.layers.GaussianNoise**: Esto es útil para mitigar el sobreajuste (podría verlo como una forma de aumento de datos aleatorio). El ruido gaussiano (GS) es una elección natural como proceso de corrupción para entradas de valor real.

Como es una capa de regularización, solo está activa en el momento del entrenamiento.

```python
np.random.seed(42)

# ENCODER
denoising_encoder = keras.models.Sequential([
    keras.layers.Flatten(input_shape=[28, 28]),
    keras.layers.GaussianNoise(0.2), # 0.2 es la std del ruido en una dist gaussiana.
    keras.layers.Dense(100, activation="selu"),
    keras.layers.Dense(30, activation="selu")
])

# DECODER
denoising_decoder = keras.models.Sequential([
    keras.layers.Dense(100, activation="selu", input_shape=[30]),
    keras.layers.Dense(28 * 28, activation="sigmoid"),
    keras.layers.Reshape([28, 28])
])

# AE
denoising_ae = keras.models.Sequential([denoising_encoder, denoising_decoder])

# COMPILE & FIT
denoising_ae.compile(loss="binary_crossentropy", optimizer=keras.optimizers.SGD(lr=1.0),
                     metrics=[rounded_accuracy])
history = denoising_ae.fit(X_train, X_train, epochs=10,
                           validation_data=(X_valid, X_valid))
```

### Funciones de Utileria (plot & display)

```python
def plot_image(image):
    plt.imshow(image, cmap="binary")
    plt.axis("off")
    
def show_reconstructions(model, images=X_valid, n_images=5):
    reconstructions = model.predict(images[:n_images])
    fig = plt.figure(figsize=(n_images * 1.5, 3))
    for image_index in range(n_images):
        plt.subplot(2, n_images, 1 + image_index)
        plot_image(images[image_index])
        plt.subplot(2, n_images, 1 + n_images + image_index)
        plot_image(reconstructions[image_index])
```

### Utilizar el Denoised Stacked AE

El AE también puede reconstruir imágenes a partir de inputs con ruido

```python
tf.random.set_seed(42)
np.random.seed(42)

noise = keras.layers.GaussianNoise(0.2)
show_reconstructions(denoising_ae, noise(X_valid, training=True))
plt.show()
```

### Entrenar con Dropout

```python
tf.random.set_seed(42)
np.random.seed(42)

# ENCODER
dropout_encoder = keras.models.Sequential([
    keras.layers.Flatten(input_shape=[28, 28]),
    keras.layers.Dropout(0.5), # otra forma de regularizacion para el entrenamiento.
    keras.layers.Dense(100, activation="selu"),
    keras.layers.Dense(30, activation="selu")
])

# DECODER
dropout_decoder = keras.models.Sequential([
    keras.layers.Dense(100, activation="selu", input_shape=[30]),
    keras.layers.Dense(28 * 28, activation="sigmoid"),
    keras.layers.Reshape([28, 28])
])

# AE
dropout_ae = keras.models.Sequential([dropout_encoder, dropout_decoder])

# COMPILE & FIT
dropout_ae.compile(loss="binary_crossentropy", optimizer=keras.optimizers.SGD(lr=1.0),
                   metrics=[rounded_accuracy])
history = dropout_ae.fit(X_train, X_train, epochs=10,
                         validation_data=(X_valid, X_valid))
```

### Utilizar el Stacked-Dropout AE

Al igual que con el GaussianNoise, un AE con dropout puede generar imagenes de input con ruido o incompletas.

```python
np.random.seed(42)

dropout = keras.layers.Dropout(0.5)
show_reconstructions(dropout_ae, dropout(X_valid, training=True))
```
